```python
import pandas as pd
```

```python
# import the data
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once per notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

# Load 2018 schoolzone speeding caught by camera


```python
speeding_camera_2018 = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2018.csv")
```

```python
speeding_camera_2018['Violation County'].unique()
```

```python
speeding_camera_2018['Violation County'] = speeding_camera_2018['Violation County'].replace({'MN' : 'NY'}, regex=True)
speeding_camera_2018['Violation County'] = speeding_camera_2018['Violation County'].replace({'BK' : 'K'}, regex=True)
```

```python
speeding_camera_2018['Violation County'].unique()
```

# Load 2019 camera speed

```python
speeding_camera_2019 = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2019.csv")
```

```python
speeding_camera_2019['Violation County'].unique()
```

```python
speeding_camera_2019['Violation County'] = speeding_camera_2019['Violation County'].replace({'MN' : 'NY'}, regex=True)
speeding_camera_2019['Violation County'] = speeding_camera_2019['Violation County'].replace({'BK' : 'K'}, regex=True)
speeding_camera_2019['Violation County'] = speeding_camera_2019['Violation County'].replace({'R' : 'ST'}, regex=True)
```

```python
speeding_camera_2019['Violation County'].unique()
```

# Load 2020 camera speed

```python
speeding_camera_2020 = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2020.csv")
```

```python
speeding_camera_2020['Violation County'].unique()
```

```python
speeding_camera_2020['Violation County'] = speeding_camera_2020['Violation County'].replace({'MN' : 'NY'}, regex=True)
speeding_camera_2020['Violation County'] = speeding_camera_2020['Violation County'].replace({'BK' : 'K'}, regex=True)
```

```python
speeding_camera_2020['Violation County'].unique()
```

# Load 2021 speed cam 

```python
speeding_camera_2021 = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2021.csv")
```

```python
speeding_camera_2021['Violation County'].unique()
```

```python
speeding_camera_2021['Violation County'] = speeding_camera_2021['Violation County'].replace({'MN' : 'NY'}, regex=True)
speeding_camera_2021['Violation County'] = speeding_camera_2021['Violation County'].replace({'BK' : 'K'}, regex=True)
```

```python
speeding_camera_2021['Violation County'].unique()
```

# Merging data sets 

```python
sc_1819 = pd.concat([speeding_camera_2018,speeding_camera_2019])
sc_181920 = pd.concat([sc_1819,speeding_camera_2020])
sc_18192021 = pd.concat([sc_181920,speeding_camera_2021])
```

```python
from google.colab import files
sc_18192021.to_csv('camera_caught_speeding_2018_2021.csv', encoding = 'utf-8-sig') 
files.download('camera_caught_speeding_2018_2021.csv')
```
