```python
# import the data
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once per notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```python
speed_cam = pd.read_csv("CameraCaught_Schoolzone_2018_21.csv")
```

```python
speed_cam['Issue Date'] = pd.to_datetime(speed_cam['Issue Date'])
speed_cam['Date'] = speed_cam['Issue Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
speed_cam['month'] = speed_cam['Issue Date'].apply(lambda x: x.strftime('%Y/-%m'))
```

```python
speed_cam_count=speed_cam.groupby(['Date']).count()[['Plate Type']]
speed_cam_count.rename(columns = {'Plate Type' : 'Num_of_Violation'}, inplace = True)
```

```python
speed_cam_count.plot()
```
