```python
red_light = pd.read_csv("red_light_summons.csv")
```

```python
red_light['VIOLATION_DATE'] = pd.to_datetime(red_light['VIOLATION_DATE'])
red_light['Date'] = red_light['VIOLATION_DATE'].apply(lambda x: x.strftime('%Y-%m-%d'))
red_light['Year'] = red_light['VIOLATION_DATE'].apply(lambda x: x.strftime('%Y'))
```

```python
boro_drivers= pd.read_csv("boro_drivers.csv")
boro_drivers.rename(columns={'boro':'CITY_NM'}, inplace=True)
boro_drivers=boro_drivers.set_index('CITY_NM')
```

```python
boro_drivers=boro_drivers.rename(index={"BK": "BROOKLYN",
                     "BX":"BRONX",
                     "MN":"MANHATTAN",
                     "QN": 'QUEENS',
                     'SI': "STATEN ISLAND"})
```

```python
# year 2018
yr_18 = ['2018']
red_light_18 = red_light['Year'].isin(yr_18)
red_light_18 = red_light.loc[red_light_18]
red_light_18['CITY_NM'] = red_light_18['CITY_NM'].replace({'NEW YORK': 'MANHATTAN'})
```

```python
df_red_light_18 = red_light_18.groupby(['CITY_NM']).count()[['CHG_LAW_CD']]
df_red_light_18.rename(columns = {'CHG_LAW_CD' : 'Num_of_Violation'}, inplace = True)
df_red_light_18 = df_red_light_18.join(boro_drivers)
df_red_light_18['Year']='2018'
df_red_light_18['Num_of_Violation'] = df_red_light_18['Num_of_Violation'].astype(float)
df_red_light_18['num_drivers'] = df_red_light_18['num_drivers'].astype(float)
df_red_light_18['Std_num_violation']=df_red_light_18['Num_of_Violation']/df_red_light_18['num_drivers']
```
